Beyond the Standard Model, two of the most famous unsolved problems in physics are:

- **The hierarchy problem**: why is gravity $10^{38}$ times weaker than electromagnetism? No

## Section 2 — Why $g = 2\pi$: Topology Fixes the Ladder Spacing

### The mystery

Section 1 assumed the hierarchy step is $g = 2\pi$. Where does that number come from?

In standard physics, there is no principle that determines the ratio between energy scales. The
ratio of the proton mass to the electron mass ($\approx 1836$) is measured, not derived. The ratio
of the electroweak scale to the Planck scale ($\approx 10^{-17}$) — the hierarchy problem — is
unexplained. Any value of $g$ would produce *some* exponential ladder, but most values would not
match the observed particle masses.

### The topological derivation

PPM derives $g$ from the symmetry structure of the fundamental process. The claim is:

> The elementary discrete symmetry of the vacuum is $\mathbb{Z}_2 \times \mathbb{Z}_2$ — the
> simplest non-trivial product of inversions. This group has four elements. The natural geometric
> space for a $\mathbb{Z}_2 \times \mathbb{Z}_2$-invariant process in 3+1 dimensions is the
> real projective space $\mathbb{RP}^3 \cong SO(3)$, which has volume $\pi^2$ in the natural
> $SU(2)$ metric.

The hierarchy step is then:
$$g^2 = |\mathbb{Z}_2 \times \mathbb{Z}_2| \times \text{Vol}(\mathbb{RP}^3) = 4 \times \pi^2 = 4\pi^2 \quad \Rightarrow \quad g = 2\pi$$

For readers without a topology background: $\mathbb{RP}^3$ is the space of all *directions* in
four-dimensional space — equivalently, the space of all possible orientations of a rigid body in 3D.
Its "size" ($\pi^2$) combined with the fourfold symmetry of the fundamental process gives the step
ratio $2\pi$ between energy levels. No particle mass data enters this calculation.

### Physical meaning of each step

Each unit increase in $k$ lowers the energy by a factor of $\sqrt{g} = \sqrt{2\pi} \approx
2.507$. This means:

- 2 steps: factor of $2\pi \approx 6.3$  
- 10 steps: factor of $(2\pi)^5 \approx 7700$ (almost four orders of magnitude)  
- 30 steps: factor of $(2\pi)^{15} \approx 3.6 \times 10^{11}$

The entire particle spectrum — thirty orders of magnitude — is covered by 75 steps of this ladder.

### What to watch

The slider varies $g$ from 5.0 to 7.8, holding everything else fixed. Five completely independent
observables are tracked simultaneously:

- **Higgs VEV** (246 GeV): fixes the EW symmetry breaking scale  
- **Top quark** (173 GeV): the heaviest known elementary particle

## Section 3 — The Electroweak Sector: $k_\text{EWSB} = 44.5$

### The mystery

Given the hierarchy ladder, the next question is: *which rung* does the electroweak sector sit on?

The Higgs mechanism is responsible for giving mass to the $W$ and $Z$ bosons and to all quarks and
leptons. It operates at the electroweak symmetry breaking (EWSB) scale — the Higgs vacuum
expectation value $v = 246.2$ GeV. The Standard Model treats this as a free parameter: it measures
$v$ experimentally and encodes it. There is no principle in the SM that predicts *why* the Higgs
mass is 125 GeV rather than 125 TeV or 125 eV. This is sometimes called the "little hierarchy
problem."

Extending the question to the lepton family: the tau is 3477 times heavier than the electron, and
the muon is 207 times heavier than the electron. These ratios are known to high precision. The SM
has no explanation for them.

### The $\mathbb{RP}^3$ emergence condition

PPM proposes that the EW sector arises when the $\mathbb{Z}_2$ projection maps $\mathbb{CP}^3
\to \mathbb{RP}^3$ at a specific level. The topology of $\mathbb{RP}^3$ (real projective
3-space, the same geometry that fixed $g$) determines the level:

$$k_\text{EWSB} = 44.5$$

The half-integer value (not an integer) reflects the $SU(2)$ doublet structure of the EW sector:
the $\mathbb{Z}_2$ identification introduces a half-period shift.

The four predictions that follow directly:

| Observable | Formula | Prediction | Observed |
|-----------|---------|-----------|---------|
| Higgs VEV | $v = 2\sqrt{2}\,(2\pi)^{1/4}\,E(44.5)$ | 246.1 GeV | 246.2 GeV |
| Top quark | $m_t = \pi\,E(44.5)$ | 172.8 GeV | 172.7 GeV |
| Tau (bare) | $E(44.5 + 3.5) = E(48)$ | 2.2 GeV | 1.777 GeV |
| Muon (bare) | $E(44.5 + 7.0) = E(51.5)$ | 88.5 MeV | 105.7 MeV |

The tau and muon are predicted at their *bare* (pre-radiative-correction) masses. The deviations
from observed values are electromagnetic self-energy corrections — calculable effects, not framework
errors. Notably, the tau, muon, and electron sit at $k = k_\text{EWSB} + n/2$ for integer $n$:
the lepton series is a $\mathbb{Z}_2$-quantized tower above the EW level.

### What to watch

The slider varies $k_\text{EWSB}$ from 43.0 to 46.0. All four predictions — Higgs VEV, top quark,

## Section 4 — The Fine Structure Constant: $\alpha = 1/137$ from Holography

### The mystery

The fine structure constant $\alpha \approx 1/137.036$ is the dimensionless measure of the
strength of the electromagnetic force. It determines how strongly electrons attract and repel,
how atoms form, how light interacts with matter. Every electromagnetic phenomenon in the universe
ultimately traces back to this one number.

Richard Feynman called it "one of the greatest damn mysteries of physics — a magic number that
comes to us with no understanding by man." It has been measured to twelve decimal places. In the
Standard Model, it is a free parameter: measured, written down, and accepted without explanation.

Why 1/137 and not 1/100 or 1/200? There is no known answer.

### The holographic derivation

PPM derives $\alpha$ from the following chain:

**Step 1 — The holographic count.** The observable universe contains $N_\text{cosmic} \approx
10^{82}$ fundamental degrees of freedom. This is the number of Planck-area cells on the cosmic
horizon (the holographic principle applied at the largest accessible scale).

**Step 2 — The effective count.** Not all $N_\text{cosmic}$ degrees of freedom are active at a
given energy level. The active count at the coherence boundary is $N_\text{eff} = N_\text{cosmic}^n$,
where $n$ is a geometric exponent. Phase space in the full $\mathbb{CP}^3$ framework has 6 real
dimensions. Under the $\mathbb{Z}_2$ projection (which reduces $\mathbb{CP}^3$ to $\mathbb{RP}^3$),
one dimension is effectively projected out. The remaining 5-dimensional holographic scaling gives
$n = 5/6$.

**Step 3 — Phase coherence.** The fine structure constant emerges from the requirement that the
two competing phase contributions are equal at the coherence boundary (explained in detail in
Section 6). This gives a sharp prediction: at $n = 5/6$, the formula reproduces $\alpha = 1/137.036$
with no adjustment.

### Why this matters ontologically

This derivation connects three completely different domains: the geometry of 6-dimensional complex
projective space ($\mathbb{CP}^3$), the holographic principle (which relates volumes to
information), and electromagnetism. If correct, it means that the strength of light is not an
arbitrary constant — it is determined by the dimensionality of the space of fundamental processes.

### What to watch

The slider varies $n$ from 0.75 to 1.0. The left panel shows the predicted $1/\alpha$ (should

The sharp feature at $n = 5/6 = 0.8\overline{3}$ is not an accident of plotting. The prediction

## Section 5 — $G$, $\Lambda$, and $H_0$ from One Holographic Count

### The two great fine-tuning problems

**The hierarchy problem.** Newton's gravitational constant $G = 6.674 \times 10^{-11}$ m³/kg·s²
sets the strength of gravity. Compared to electromagnetism, gravity is weaker by a factor of
$10^{36}$. Why? No principle in physics explains this ratio. The Standard Model simply
measures $G$ and inserts it.

**The cosmological constant problem.** Quantum field theory predicts that empty space should carry
a vacuum energy density, contributed by quantum fluctuations of every field. Summing these
contributions gives a value for the cosmological constant $\Lambda$ of order $10^{70}$ m$^{-2}$.
The observed value is $\Lambda \approx 1.1 \times 10^{-52}$ m$^{-2}$ — smaller by a factor of
$10^{122}$. This is routinely called the worst numerical prediction in the history of science.
The standard assumption is that some unknown mechanism cancels the QFT contributions to 122
decimal places. No such mechanism has been found.

### The framework resolution

Both constants involve the same holographic count $N_\text{cosmic}$ — but with different
exponents:

$$G = \frac{16\pi^4 \hbar c\,\alpha}{m_\pi^2 \,\sqrt{N_\text{cosmic}}} \propto N^{-1/2}$$

$$\Lambda = \frac{2(m_\pi c^2)^2}{(\hbar c)^2\, N_\text{cosmic}} \propto N^{-1}$$

The key insight is the **different exponents**: $G$ falls as $N^{-1/2}$ while $\Lambda$ falls as
$N^{-1}$. Because they have genuinely different slopes on a log-log plot, only *one* value of
$N_\text{cosmic}$ can make both match observation simultaneously. That value is $N = 10^{82}$.

Gravity is weak not because of a fine-tuning, but because the universe is old and large. The
holographic count $N_\text{cosmic}$ grows with the size of the cosmic horizon, and $G \propto
N^{-1/2}$ means that at late cosmic times, gravity becomes diluted. The cosmological constant is
small for the same reason: $\Lambda \propto N^{-1}$ falls even faster. These are not adjustments
— they are consequences of the universe's age.

### The Hubble constant connection

The Hubble constant $H_0$ describes how fast the universe is currently expanding:
$$H_0^\text{dS} = c\sqrt{\frac{\Lambda}{3}} \propto N^{-1/2}$$

This is the de Sitter (dark-energy-dominated) contribution to $H_0$. It shares the $N^{-1/2}$
scaling of $G$ but derives from $\Lambda$ through a completely different physical mechanism.
At $N = 10^{82}$, this gives $H_0^\text{dS} \approx 54$ km/s/Mpc — the observed $67.4$
km/s/Mpc includes an additional contribution from the matter density of the universe, which
accounts for the remaining $\sim 20$%. The framework captures the dominant dark-energy term;
matter corrections require the full Friedmann equation.

## Section 6 — Phase Coherence: Two Exponentials Crossing at Life

### Phase accumulation and coherence

Every physical process accumulates phase as it propagates. In quantum mechanics, this is the
complex phase of the wavefunction. In thermodynamics, phase is related to entropy — the number of
accessible states. At each rung $k$ of the hierarchy, there are two competing phase contributions:

$$\Phi_\text{thermal}(k) = N_\text{eff} \cdot g^{(51-k)/2} \qquad \text{(thermal, decreasing in } k\text{)}$$

$$\Phi_\text{quantum}(k) = \alpha \cdot g^k \qquad \text{(quantum Berry phase, increasing in } k\text{)}$$

$\Phi_\text{thermal}$ is the effective number of thermal degrees of freedom at level $k$. It
starts large at low $k$ (high energy, many thermal fluctuations) and decreases as $k$ increases
(lower energy, fewer thermal modes). $\Phi_\text{quantum}$ is the electromagnetic Berry phase
accumulated by a quantum process cycling through $k$ energy levels. It starts small at $k = 0$
(few levels, little phase) and grows exponentially.

These two functions are **monotonically opposite in direction**. They must cross exactly once.

### The crossing point and $n = 5/6$

The crossing occurs at $k_c$ where $\Phi_\text{thermal}(k_c) = \Phi_\text{quantum}(k_c)$.
Solving analytically:

$$k_\text{cross} = \frac{2}{3} \left( \frac{\ln(N_\text{eff}/\alpha)}{\ln g} + 25.5 \right)$$

The temperature corresponding to the crossing level is $T = E(k_\text{cross})/k_B$.

The slider varies $n$ in $N_\text{eff} = N_\text{cosmic}^n$, which shifts the thermal curve
vertically on the log-scale plot, moving the crossing point. At $n = 5/6$ — the same value that
Section 4 established from the requirement $\alpha = 1/137$ — the crossing falls at
$k_\text{cross} \approx 75.354$, corresponding to $T = 310$ K.

This is not a separate prediction. The value of $n$ was already determined by $\alpha$. The
temperature $310$ K is an *automatic consequence* of that determination.

### What to watch

The two curves on the left panel are exponentials in $k$: one decreasing (blue, thermal) and one
increasing (red, quantum). Drag the slider. As $n$ changes, the blue curve shifts up or down,
moving the crossing point left or right. The table shows the exact crossing $k$-value and the
equivalent temperature. Only at $n = 5/6$ does the temperature read $310$ K.
